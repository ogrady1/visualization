```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL ##install as pillow if needed
import requests

np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0,100,50)
y2 = np.random.randint(0, 100, 50)
```

```python
#plot both y variables on the same axis

fig, ax = plt.subplots(figsize=(5, 3))
ax.hist(y1)
ax.hist(y2)
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y1, label='person 1')
ax.plot(x, y2, label='person 2')
ax.legend(loc= 'lower right')
```

```python
#styling legend
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y1, label='person 1')
ax.plot(x, y2, label='person 2')
ax.legend(loc= 'lower right',
          frameon = True,
          fontsize = 12,
          ncol = 2,
          shadow = True,)
          
```

```python
#moving legend outside the plot. last two lines are important to adjust

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y1, label='Person 1')
ax.plot(x, y2, label='Person 2')
ax.legend(loc='lower left', 
          bbox_to_anchor=(1, 1)) 
```

Text and Annotations

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y1, label='Person 1')
ax.plot(x, y2, label='Person 2')
ax.legend(loc='lower right')

#adding text and modifying style
ax.text(10, 95, 'this value is important!',
        ha = 'center',
        color = 'red',
        fontsize = 20)

```

```python
#testing specific locations, useful when you want the text to follow the data when it changes

fig, ax = plt.subplots()
ax.axis([0,10,0,10])


ax.text(1, 5, " . Data: (1,5)", transform = ax.transData),
ax.text(0.5,0.1, " . axes:(0.5, 0.1)", transform = ax.transAxes),
ax.text(0.2, 0.2, " . figure: (0.2,0.2)", transform = fig.transFigure),
```

```python
#adding arrow

fix, ax= plt.subplots(figsize=(5, 3))
ax.scatter(x, y1, label='Person 1')
ax.scatter(x, y2, label='Person 2')
ax.legend(loc='lower right')

ax.annotate('This is important',
            xy=(10, 95),
            xytext=(20, 94),
            arrowprops=dict(facecolor='black'))
```

```python
#styling arrow

fix, ax= plt.subplots(figsize=(5, 3))
ax.scatter(x, y1, label='Person 1')
ax.scatter(x, y2, label='Person 2')
ax.legend(loc='lower right')

ax.annotate('This is important',
            xy=(10, 95),
            xytext=(20, 94),
            arrowprops=dict(arrowstyle = "wedge", facecolor='hotpink'))
```

```python
#removing axis labels 


fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x, y1, label='Person 1')
ax.scatter(x, y2, label='Person 2')
ax.legend(loc='lower right')

ax.yaxis.set_major_locator(plt.NullLocator())
ax.xaxis.set_major_formatter(plt.NullFormatter())
```

```python
#adjusting the x-axis ticks


fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x, y1, label='Person 1')
ax.scatter(x, y2, label='Person 2')
ax.legend(loc='lower right')


ax.xaxis.set_major_locator(plt.MaxNLocator(3)) #maximum of 3 ticks on the x-axis
ax.yaxis.set_major_locator(plt.MultipleLocator(5)) #intervals of 5 ticks on the y-axis   
```

```python
#styling the ticks more

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x, y1, label='Person 1')
ax.scatter(x, y2, label='Person 2')
ax.legend(loc='lower right')


ax.xaxis.set_major_locator(plt.MaxNLocator(3)) 
ax.yaxis.set_major_locator(plt.MultipleLocator(5))  

plt.xticks(rotation=45, ha='right', 
           fontsize=12, 
           color='blue', 
           fontweight='bold')
```

```python
#changing axis fonts using dictionary

font1 = {'family': 'serif', 'weight': 'bold', 'size': 14,}

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x, y1, label='Person 1')
ax.scatter(x, y2, label='Person 2')
ax.legend(loc='lower right')

ax.set_xlabel('Shiny new label', fontdict=font1, color='blue') #color has to be here, cant be part of dictionary

ax.xaxis.set_major_locator(plt.MaxNLocator(3)) #maximum of 3 ticks on the x-axis
ax.yaxis.set_major_locator(plt.MultipleLocator(5)) #intervals of 5 ticks on the y-axis   
```

```python
#style presets

plt.style.use('ggplot')

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x, y1)
ax.scatter(x, y2)
```

```python
np.random.seed(613)
x1 = np.arange(50)
y1 = np.random.randint(0,75,50)
x2 = np.array(["luffy", "Zono", "Nami", "Ussop", "Sanji"])
y2 = np.array([110,180,240,99,220])

```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2,
                               nrows=1,
                               figsize=(7, 3))
ax1.scatter(x1, y1)
ax2.bar(x2, y2)
```

```python
#practice custumozing 

fig, (ax1, ax2) = plt.subplots(ncols=2,
                               nrows=1,
                               figsize=(9, 3))


font1 = {'family': 'serif', 'weight': 'bold', 'size': 14,}  
font2 = {'family': 'sans-serif', 'size': 10,}

ax1.scatter(x1, y1,
            marker = 'o',
            color = 'blue',)
ax2.bar(x2, y2, color='orange')

ax1.set_title('Scatter plot', fontdict=font1, color='blue')
ax1.set_xlabel('X-axis label', fontdict=font1, color='blue')
ax1.set_ylabel('Y-axis label', fontdict=font1, color='blue')
ax2.set_title('Bar plot', fontdict=font2, color='orange')
ax2.set_xlabel('X-axis label', fontdict=font2, color='orange')
ax2.set_ylabel('Y-axis label', fontdict=font2, color='orange')
```

```python
#subplots without a grid
plt.style.use('default') #getting rid of the previous style, only needed if it was used before
fig, someaxes = plt.subplot_mosaic([['ax1','ax3'], 
                                    ['ax2','ax3']],
                                    figsize=(7, 4))

someaxes ['ax1'].scatter(x1, y1)
someaxes ['ax2'].bar(x2, y2)
someaxes ['ax3'].plot(x, y1)
```

```python
#adding labels 
plt.style.use('default') 
fig, someaxes = plt.subplot_mosaic([['ax1','ax3'], 
                                    ['ax2','ax3']],
                                    figsize=(7, 4),
                                    layout='constrained') #this is important to make sure the labels fit


someaxes ['ax1'].scatter(x1, y1)
someaxes ['ax2'].bar(x2, y2)
someaxes ['ax3'].plot(x, y1)

someaxes ['ax1'].set_xlabel('A Big Label', fontsize = 18)
someaxes ['ax2'].set_xlabel('Another Label', fontsize = 18)
someaxes ['ax3'].set_xlabel('Label 2: 2 Fast 2 Furious', fontsize = 18)
```
