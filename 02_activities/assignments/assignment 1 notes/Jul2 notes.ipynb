```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Set random seed for reproducibility
np.random.seed(613)

# Generate random data
x = np.arange(50)
y = np.random.randint(0, 100, 50)

# Create a figure and axis
fig, ax = plt.subplots(figsize=(5, 3))
#ax.scatter(x,y)
#ax.bar (x,y)
#ax.barh(x,y)
#ax.plot(x,y)
ax.hist(y)
```

```python
#add font libraries/styles
font1 = {'family': 'sans-serif', 'color':  'blue','size': 20}
font2 = {'family': 'monospace', 'color':  'green', 'size': 14}

# Set titles and labels with custom fonts
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
ax.set_title('total growth over time', fontdict=font1, loc='left')
ax.set_ylabel('total growth', fontdict=font2)
ax.set_xlabel('years since start', fontdict=font2)

```

```python
#add custom markers and colors
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x, y, marker = '*', color = 'indigo')
```

```python
#add custom markers and colors with edge and face colors
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y, marker = '*', color = '#7425B9', linestyle= '--', linewidth= 2, markersize= 12, markeredgecolor= '#fa9359', markerfacecolor= '#000000')
ax.grid()
```

```python
# libraries
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.random.normal(size=50000)
y = x * 3 + np.random.normal(size=50000)

# A histogram 2D
plt.hist2d(x, y, bins=(50, 50), cmap=plt.cm.Reds)

# Add a basic title
plt.title("A 2D histogram")

# Show the graph
plt.show()
```
